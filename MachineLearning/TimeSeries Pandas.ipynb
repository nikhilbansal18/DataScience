```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

usb=CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
## Creating custom holiday calendar
# nearest_workday - if public holiday falls on weekend; give on nearest weekday

from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday

class myCalendar(AbstractHolidayCalendar):
    """
        My custom calendar
    """

    rules = [
        Holiday("XYZ Day", month=7, day=9, observance=nearest_workday),
    ]

mch=CustomBusinessDay(calendar=myCalendar())
```

```python
rng = pd.date_range(start="07/01/2017",end="07/21/2017",freq=mch)
rng
```

```python
## Shift by 2 days
fb.shift(2)
```

```python
# shift to left
fb.shift(-1)
```

```python
fb['Prev Day Price'] = fb['Price'].shift(1)
fb.head()
```

```python
fb['1 Day Change'] = fb['Price'] - fb['Prev Day Price']
fb.head()
```

```python
fb['5 day % return'] = (fb['Price'] - fb['Price'].shift(5))*100/fb['Price'].shift(5)
fb
```
